### Question3

```python
import pandas as pd
import time
```

```python
options = wb.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
driver = wb.Chrome("C:/Users/smhrd/시험/chromedriver.exe", options = options)
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(0.5)
titleList = []
priceList = []
for i in range(10) :
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    title = driver.find_element_by_css_selector(".itemtit")
    price = driver.find_element_by_css_selector(".price_real")
    # 순수한 글자정보만 저장해주는 리스트 작업
    titleList.append(title.text)
    priceList.append(price.text)
    driver.back()
    time.sleep(1)
dic = {"상품명" : titleList, "가격" : priceList}
gmarket = pd.DataFrame(dic)
gmarket

# 데이터프레임 제작
# csv형식으로 저장
```

### Question2

```python
url = 'https://www.melon.com/chart/index.htm'
res = req.get(url,headers = h)
res
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
title = soup.select('div.ellipsis.rank01 > span > a')
title
```

```python
len(title)
```

```python
singer = soup.select('div.ellipsis.rank02 > a:nth-child(1)')
```

```python
len(singer)
```

```python
title_list = []
singer_list = []
rank_list = []

for i in range(len(title)):
    title_list.append(title[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
```

```python
dic = {'노래제목':title_list, '가수' : singer_list,'순위' : rank_list}
df=pd.DataFrame(dic)
df.set_index('순위', inplace = True)
df
```
